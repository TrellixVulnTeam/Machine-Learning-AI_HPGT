# Similar to the conditional expression, the isin() conditional function returns a True for each row the values are in the provided list. To filter the rows based on such a function, use the conditional function inside the selection brackets []. In this case, the condition inside the selection brackets df["Pclass"].isin([2, 3]) checks for which rows the Pclass column is either 2 or 3.

```python
# Only class 1
pclass1 = df[df['Pclass'] == 1]
pclass1
```

```python
# Only class 2
pclass2 = df[df['Pclass'] == 2]
pclass2
```

```python
# Combining the two, Pclass 1 and Pclass 2 using different syntax
combined_12 = df[(df['Pclass'] == 1) | (df['Pclass'] == 2)]
```

```python
combined_12
```

# Note: When combining multiple conditional statements, each condition must be surrounded by parentheses (). Moreover, you can not use or/and but need to use the or operator | and the and operator &.

```python
# Total number of people who survived out of the class 1 and class 2
combined_12.Survived.sum()
```

```python
#To select rows based on a conditional expression, 
# use a condition inside the selection brackets [].

# The condition inside the selection brackets df["Age"] > 45 
# checks for which rows the Age column has a value larger than 45:
greater_45 = df['Age'] > 45
greater_45
```

# How to select specific rows and columns from a DataFrame

In this case, a subset of both rows and columns is made in one go and just using selection brackets [] is not sufficient anymore. The loc/iloc operators are required in front of the selection brackets []. When using loc/iloc, the part before the comma is the rows you want, and the part after the comma is the columns you want to select.

When using the column names, row labels or a condition expression, use the loc operator in front of the selection brackets []. For both the part before and after the comma, you can use a single label, a list of labels, a slice of labels, a conditional expression or a colon. Using a colon specifies you want to select all rows or columns.

```python
# Iâ€™m interested in rows 10 till 25 and columns 3 to 5.
df.iloc[9:25, 2:5]
```

When selecting specific rows and/or columns with loc or iloc, new values can be assigned to the selected data. For example, to assign the name Archimedes to the first 3 elements of the third column:

```python
df
```

## TIME SERIES

# This is a powerful and efficient functionality for performing resampling operations during conversion. This is extremely common in, but not limited to, financial applications

```python
# Parsing time series info from various sources and formats
import datetime

date_time = pd.to_datetime(['1/1/2018', np.datetime64('2018-01-01'), datetime.datetime(2018, 1, 1)])

date_time
```

```python
# Generate sequences of fixed-frequency dates and time spans
date_time_ = pd.date_range('2018-01-01', periods=3, freq='H')
date_time_ 
```

```python
# Manipulating and converting date times with timezone information
date_time_ = date_time_.tz_localize('UTC')
date_time_
```

```python
date_time_.tz_convert('US/Pacific')
```

```python
# Resampling or converting a time series to a particular frequency
idx = pd.date_range('2018-01-01', periods=5, freq='H')

ts = pd.Series(range(len(idx)), index=idx)

ts
```

```python
ts.resample('2H').mean()
```

```python
# Performing date and time arithmetic with absolute or relative time increments
friday = pd.Timestamp('2018-01-05')

friday.day_name()
```

```python
# Add a day
saturday = friday + pd.Timedelta('1 day')
saturday.day_name()
```

```python
# Add pme business day (Friday -> Monday)
monday = friday + pd.offsets.BDay()
monday.day_name()
```

# pandas captures 4 general time related concepts:

1. Date times: A specific date and time with timezone support. Similar to datetime.datetime from the standard library.

2. Time deltas: An absolute time duration. Similar to datetime.timedelta from the standard library.

3. Time spans: A span of time defined by a point in time and its associated frequency.

4. Date offsets: A relative time duration that respects calendar arithmetic. Similar to dateutil.relativedelta.

relativedelta from the dateutil package.

For time series data, itâ€™s conventional to represent the time component in the index of a Series or DataFrame so manipulations can be performed with respect to the time element.

```python
pd.Series(range(3), index=pd.date_range('2000', freq='D', periods=3))
```

```python
# Series and DataFrame can directly also support the time component as data itself.
pd.Series(pd.date_range('2000', freq='D', periods=3))
```

Series and DataFrame have extended data type support and functionality for datetime, timedelta and Period data when passed into those constructors. DateOffset data however will be stored as object data.

```python
pd.Series(pd.period_range('1/1/2011', freq='M', periods=4))
```

```python
pd.Series([pd.DateOffset(1), pd.DateOffset(2)])
```

```python
pd.Series(pd.date_range("1/1/2011", freq="M", periods=3)) 
```
